### Data Description

- Customer - Customer ID

- ST - State where customers live

- Gender - Gender of the customer

- Education - Background education of customers 

- Customer Lifetime Value - Customer lifetime value(CLV) is the total revenue the client will derive from their entire relationship with a customer. In other words, is the predicted or calculated value of a customer over their entire duration as a policyholder with the insurance company. It is an estimation of the net profit that the insurance company expects to generate from a customer throughout their relationship with the company. Customer Lifetime Value takes into account factors such as the duration of the customer's policy, premium payments, claim history, renewal likelihood, and potential additional services or products the customer may purchase. It helps insurers assess the long-term profitability and value associated with retaining a particular customer.

- Income - Customers income

- Monthly Premium Auto - Amount of money the customer pays on a monthly basis as a premium for their auto insurance coverage. It represents the recurring cost that the insured person must pay to maintain their insurance policy and receive coverage for potential damages, accidents, or other covered events related to their vehicle.

- Number of Open Complaints - Number of complaints the customer opened

- Policy Type - There are three type of policies in car insurance (Corporate Auto, Personal Auto, and Special Auto)

- Vehicle Class - Type of vehicle classes that customers have Two-Door Car, Four-Door Car SUV, Luxury SUV, Sports Car, and Luxury Car

- Total Claim Amount - the sum of all claims made by the customer. It represents the total monetary value of all approved claims for incidents such as accidents, theft, vandalism, or other covered events.

```python
# Your code here

import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
data = pd.read_csv (url)

# Dimensions
print("Dataset dimensions:", data.shape)
```

```python
# Data types
  # Convert problematic columns
data['Customer Lifetime Value'] = (
    data['Customer Lifetime Value']
    .astype(str)
    .str.replace('%', '', regex=False)
    .replace(['nan', 'NaN', 'None', ''], pd.NA)
)
data['Customer Lifetime Value'] = pd.to_numeric(data['Customer Lifetime Value'], errors='coerce')

data['Number of Open Complaints'] = (
    data['Number of Open Complaints']
    .astype(str)
    .str.extract(r'(\d+)')
)
data['Number of Open Complaints'] = pd.to_numeric(data['Number of Open Complaints'], errors='coerce')

  # Display data types
print("Data Types:")
print(data.dtypes)
print("-" * 60)

print("Fix Suggestions:")
print("""
- Convert 'Customer Lifetime Value' to numeric 
- Convert 'Number of Open Complaints' to numeric 
- Standardize 'ST' (state), 'GENDER', and 'Education' spelling
""")
```

```python
# Unique values per column
unique_counts = data.nunique()
print("\n Number of Unique Values per Column:")
print(unique_counts)
print("-" * 60)

  # Identify categorical and numerical columns
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
numerical_cols = data.select_dtypes(include=['number']).columns.tolist()

print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
print("-" * 60)
```

```python
#Categorical Column Analysis

print("\nCategorical Column Summary (top unique values):")
for col in categorical_cols:
    print(f"\n{col} (Unique: {data[col].nunique()})")
    print(data[col].value_counts().head(10))

print("-" * 60)
```

```python
#Numerical Column Analysis

# Basic statistics
num_summary = data[numerical_cols].describe().T
num_summary["median"] = data[numerical_cols].median()
num_summary["mode"] = data[numerical_cols].mode().iloc[0]
num_summary["std_dev"] = data[numerical_cols].std()

print("\nNumerical Summary Statistics:")
print(num_summary)
print("-" * 60)

# Range of values
print("\nRange of Each Numerical Column:")
for col in numerical_cols:
    print(f"{col}: {data[col].min()} â†’ {data[col].max()}")

print("-" * 60)

# Insights
print("ðŸ’¡ Insights:")
print("""
- 'Customer Lifetime Value' and 'Monthly Premium Auto' have very large ranges â†’ check for outliers.
- 'Number of Open Complaints' has only one unique value â†’ possible data issue.
- 'Income' and 'Total Claim Amount' look reasonable for customer-level data.
""")
```

```python
 #Summary for Categorical Columns

print("\nSummary Statistics for Categorical Columns:")
for col in categorical_cols:
    print(f"\n {col}")
    print("Unique Values:", list(data[col].dropna().unique()))
    print("Most Frequent Value(s):")
    print(data[col].value_counts().head(3))
    print("-" * 40)
```

### Exercise 1

The marketing team wants to know the top 5 less common customer locations. Create a pandas Series object that contains the customer locations and their frequencies, and then retrieve the top 5 less common locations in ascending order.

```python
# Your code here

import pandas as pd

# Create a pandas Series of customer locations and their frequencies
location_freq = data['ST'].value_counts()

# Retrieve the 5 least common locations in ascending order
least_common_5 = location_freq.sort_values(ascending=True).head(5)

# Display results
print("Top 5 Least Common Customer Locations:\n")
print(least_common_5)
```

### Exercise 2

The sales team wants to know the total number of policies sold for each type of policy. Create a pandas Series object that contains the policy types and their total number of policies sold, and then retrieve the policy type with the highest number of policies sold.

*Hint:*
- *Using value_counts() method simplifies this analysis.*
- *Futhermore, there is a method that returns the index of the maximum value in a column or row.*


```python
# Your code here

# Create a Series showing total number of policies sold per type
policy_counts = data['Policy Type'].value_counts()

# Retrieve the policy type with the highest number of policies sold
top_policy_type = policy_counts.idxmax()
top_policy_count = policy_counts.max()

# Display the results
print("Total Policies Sold by Type:\n")
print(policy_counts)
print("\nPolicy Type with the Highest Sales:")
print(f"{top_policy_type} â†’ {top_policy_count} policies sold")
```

### Exercise 3

The sales team wants to know if customers with Personal Auto have a lower income than those with Corporate Auto. How does the average income compare between the two policy types?

- Use *loc* to create two dataframes: one containing only Personal Auto policies and one containing only Corporate Auto policies.
- Calculate the average income for each policy.
- Print the results.

```python
# Your code here

import pandas as pd

# Create separate DataFrames for each policy type using .loc

personal_auto = data.loc[data['Policy Type'] == 'Personal Auto']
corporate_auto = data.loc[data['Policy Type'] == 'Corporate Auto']

# Calculate average income for each policy type

avg_income_personal = personal_auto['Income'].mean()
avg_income_corporate = corporate_auto['Income'].mean()

# Display the results

print("Average Income by Policy Type:\n")
print(f"Personal Auto:  ${avg_income_personal:,.2f}")
print(f"Corporate Auto: ${avg_income_corporate:,.2f}")

# Optional: Simple conclusion
if avg_income_personal < avg_income_corporate:
    print("\nCustomers with Personal Auto policies have a LOWER average income.")
else:
    print("\nCustomers with Personal Auto policies have a HIGHER average income.")
```
